Loss function:

- Linear Regression loss function

$ J( \theta ) = \frac{1}{m} -\sum{(y_i - yhat_i)^2} $

- Logisitic regression loss function

$ J( \theta ) = \frac{1}{m} -\sum{y_ilog(y_i)(1-y_i)log(1-yhat_i)}  $



```python
def loss_function(yhat, y):
    m = len(y)
    loss = -np.sum((y - yhat)**2)
    return loss / m
```

```python
# Classification score function 
def score_function(yhat, y):
    return np.where(yhat == y, 1, 0)/len(y)
```
